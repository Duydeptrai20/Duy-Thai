```
import numpy as np
import skfuzzy as fuzz 
import matplotlib.pyplot as plt
# tạo khoảng mờ 
x = np.arange(30,81,1)
# tạo hàm liên thuộc 
slow = fuzz.trimf(x,[30,30,50])
medium = fuzz.trimf(x,[50,60,70])
medium_fast = fuzz.trimf(x,[60,70,80])
medium_speed = fuzz.trimf(x,[70,80,80])
# vẽ ham liên thuộc 
plt.figure()
plt.plot(x,slow,'b',linewidth = 1.5, label='Slow')
plt.plot(x,medium,'k',linewidth = 1.5, label='Medium')
plt.plot(x,medium_fast,'m',linewidth = 1.5, label='Medium_fast')
plt.plot(x,medium_speed,'v',linewidth = 1.5, label='Medium_speed')
plt.title('Hàm liên thuộc của tốc độ')
plt.xlabel('khoảng mờ') 
plt.ylabel('Giá trị của hàm liên thuộc')
plt.legend(loc='center right',bbox_to_anchor=(1.25,0.5),ncol=1,fancybox=True, shadow=True)
```

```
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
#Tạp biến ra vào
food = ctrl.Antecedent(np.arange(0,10.1,0.1),'food')#thay đổi độ phân giải
service = ctrl.Antecedent(np.arange(0, 10.1, 0.1),'service')
tip = ctrl.Consequent(np.arange(10,20.1,0.1),'tip')
 #Tạo hàm liên thuộc tự động
food.automf(3)
service.automf(3)
#food.view()
#service.view()
# Tạo hàm liên thuộc bằng tay
tip['less'] = fuzz.trimf(tip.universe, [10,12.5, 15])
tip['medium'] = fuzz.trimf(tip.universe, [12.5, 15, 17.5])
tip['much'] = fuzz.trimf(tip.universe, [15, 17.5, 20])
#tip.view()
rule1 = ctrl.Rule((food['poor'] & service['poor']) | (food['poor'] & service['average'])|(food['average'] & service['poor']), tip['less'])
rule2 = ctrl.Rule((food['poor']& service['good'])|(food['good']&service['poor'])|(food['average']&service['average']),tip['medium'])
rule3 = ctrl.Rule((food['good']&service['average'])|(food['good']&service['good'])|(food['average']&service['good']), tip['much'])

tipping_ctrl = ctrl.ControlSystem([rule1,rule2,rule3])
tipping=ctrl.ControlSystemSimulation(tipping_ctrl)
tipping.input['food']=1
tipping.input['service']=1 
tipping.compute()
print(tipping.output['tip'])
tip.view(sim=tipping) 
```

```
from matplotlib import checkdep_dvipng
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
x= np.arange(10,101,1)
y= np.arange(0,1.1,0.1)
z= np.arange(10,201,1)

#Tạo biến ra vào:
rice = ctrl.Antecedent(np.arange(0,10.1,0.1),'rice')
time = ctrl.Antecedent(np.arange(10,201,1),'time')
power = ctrl.Consequent(np.arange(10,101,1),'power') 

# tạo hàm liên thuộc:
rice['less'] = fuzz.gaussmf(rice.universe,1,2)
rice['medium'] = fuzz.gaussmf(rice.universe,4.5,2)
rice['much'] = fuzz.gaussmf(rice.universe,7.5,2)
time['short'] = fuzz.gaussmf(time.universe,50,50)
time['medium'] = fuzz.gaussmf(time.universe,110,50)
time['long'] = fuzz.gaussmf(time.universe,170,50)
time['verylong'] = fuzz.gaussmf(time.universe,200,50)
power['verylow']= fuzz.gaussmf(power.universe,10,5)
power['low']=fuzz.gaussmf(power.universe,20,5)
power['medium']=fuzz.gaussmf(power.universe,30,5)
power['high']=fuzz.gaussmf(power.universe,40,10)
power['veryhigh']=fuzz.gaussmf(power.universe,50,5)
rice.view()
time.view()
power.view()
rule1 = ctrl.Rule(rice['less']& time['short'],power['medium'])
rule2 = ctrl.Rule(rice['less']& time['medium'],power['low'])
rule3 = ctrl.Rule(rice['less']& time['long'],power['verylow'])
rule4 = ctrl.Rule(rice['less']& time['verylong'],power['low'])
rule5 = ctrl.Rule(rice['medium']& time['short'],power['high'])
rule6 = ctrl.Rule(rice['medium']& time['medium'],power['medium'])
rule7 = ctrl.Rule(rice['medium']& time['long'],power['low'])
rule8 = ctrl.Rule(rice['medium']& time['verylong'],power['low'])
rule9 = ctrl.Rule(rice['medium']& time['short'],power['veryhigh'])
rule10 = ctrl.Rule(rice['much']& time['medium'],power['high'])
rule11 = ctrl.Rule(rice['much']& time['long'],power['medium'])
rule12 = ctrl.Rule(rice['much']& time['verylong'],power['medium'])

tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7,rule8, rule9, rule10, rule11, rule12])
tipping= ctrl.ControlSystemSimulation(tipping_ctrl)
tipping.input['rice'] = 1
tipping.input['time'] = 30
tipping.compute()
print(tipping.output['power'])
power.view(sim=tipping)  # sim: simulation

```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
#tạo dữ liệu
data = pd.read_csv('Position_Salaries.csv')
data.hist()
plt.show()
correlations = data.corr()
print(correlations)

x = data['Level']
y = data['Salary']
#sns.pairplot(data)
#hiển thị dữ liệu
plt.plot(x,y,'r.',label='Data')
from sklearn import linear_model
#hồi quy đa thức
poly_model = linear_model.LinearRegression()
#Tạo ra khoảng dữ liệu
x = pd.DataFrame([data['Level'],data['Level']**2,data['Level']**3,data['Level']**4,data['Level']**5,data['Level']**6,data['Level']**7,data['Level']**8,data['Level']**9,data['Level']**10]).T
#Huấn luyện mô hình 
poly_model.fit(x,y=data['Salary'])
#Các giá trị hàm hồi quy 
print('Giá trị intercept là:',poly_model.intercept_)
print('Giá trị coeffiets là:',poly_model.coef_)
print('Độ chính xác:',poly_model.score(x,y=data['Salary']))
#Tạo khoảng dữ liệu
poly_range = np.arange(1,10,0.1)
#Giá trị tiên đoán
poly_pred = pd.DataFrame([poly_range,poly_range**2,poly_range**3,poly_range**4,poly_range**5,poly_range**6,poly_range**7,poly_range**8,poly_range**9,poly_range**10]).T
#Giá trị đáp ứng
y_value = poly_model.predict( poly_pred)
data.plot(kind = 'scatter', x ='Level', y = 'Salary', figsize = (9,9), color = 'black', xlim = (0,10))
plt.plot(poly_range, y_value, color = 'blue')
```
